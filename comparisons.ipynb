```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv('Lab2_Results_worst_case.csv')

plt.figure(figsize=(24,12))
ax = sb.lineplot(x=data["Number of Vertices"], y=data["Time Q1 (ms)"], color='blue', label='Q1')
ax = sb.lineplot(x=data["Number of Vertices"], y=data["Time Q2 (ms)"], color='red', label='Q2')
ax.set(ylabel="Time (ms)")
```

```python
avgCase = pd.read_csv('Lab2_Results_average_case.csv')
worstCase = pd.read_csv('Lab2_Results_worst_case.csv')
bestCase = pd.read_csv('Lab2_Results_best_case.csv')

avgCase = avgCase[:500]
worstCase = worstCase[:500]
bestCase = bestCase[:500]

plt.figure(figsize=(24,12))
ax = sb.lineplot(x=avgCase["Number of Vertices"], y=avgCase["Time Q1 (ms)"], color='orange', label='Q1 Avg Case')
ax = sb.lineplot(x=avgCase["Number of Vertices"], y=avgCase["Time Q2 (ms)"], color='blue', label='Q2 Avg Case')
ax = sb.lineplot(x=worstCase["Number of Vertices"], y=worstCase["Time Q1 (ms)"], color='orange', label='Q1 Worst Case')
ax = sb.lineplot(x=worstCase["Number of Vertices"], y=worstCase["Time Q2 (ms)"], color='blue', label='Q2 Worst Case')
ax = sb.lineplot(x=bestCase["Number of Vertices"], y=bestCase["Time Q1 (ms)"], color='orange', label='Q1 Best Case')
ax = sb.lineplot(x=bestCase["Number of Vertices"], y=bestCase["Time Q2 (ms)"], color='blue', label='Q2 Best Case')
ax.set(ylabel="Time (ms)")
```
