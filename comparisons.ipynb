```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv('Lab2_Results.csv')

plt.figure(figsize=(24,12))
ax = sb.lineplot(x=data["Number of Vertices"], y=data["Time Q1 (ms)"], color='blue', label='Q1')
ax = sb.lineplot(x=data["Number of Vertices"], y=data["Time Q2 (ms)"], color='red', label='Q2')
ax.set(ylabel="Time (ms)")
```
