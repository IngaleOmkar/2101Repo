```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
data = pd.read_csv('Time_S.csv')
```

```python
data
```

```python
plt.figure(figsize=(24,12))
ax = sb.lineplot(x=data["s"], y=data["MergeSort"], color='blue', label='mergeSort')
ax = sb.lineplot(x=data["s"], y=data["HybridSort"], color='red', label='hybridSort')
ax.set(ylabel="Time (ms)")
```

```python
plt.figure(figsize=(24,12))
ax = sb.lineplot(x=data["s"], y=data["MergeSort"]-data["HybridSort"], color='blue', label='Merge - Hybrid')
ax.set(ylabel="Difference in Time (ms)")
```
