```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
def plotGraph(file1):
    plt.figure(figsize=(24,12))
    ax = sb.lineplot(x=file1["Array Size"], y=file1["mergeSort Time (ms)"], color='blue', label='mergeSort')
    ax = sb.lineplot(x=file1["Array Size"], y=file1["insertionSort Time (ms)"], color='orange', label='insertionSort')
    ax = sb.lineplot(x=file1["Array Size"], y=file1["hybridSort Time (ms)"], color='green', label='hybridSort')
    ax.set(ylabel="Time (ms)")
    
file4 = 'results_combined.csv'
```

```python
combined = pd.read_csv(file4)
```

```python
combined
```

```python
combined.loc[0,'hybridSort Time (ms)'] = 0
combined.loc[0,'insertionSort Time (ms)'] = 0
combined.loc[0,'mergeSort Time (ms)'] = 0
```

```python
combined
```

```python
plotGraph(combined[:200])
```
