```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
data = pd.read_csv('data_43.csv')
```

```python
data
```

```python
plt.figure(figsize=(24,12))
ax = sb.lineplot(x=data["ArraySize"], y=data["MergeSort"], color='blue', label='mergeSort')
ax = sb.lineplot(x=data["ArraySize"], y=data["HybridSort"], color='red', label='hybridSort')
ax.set(ylabel="Time (ms)")
```

```python
plt.figure(figsize=(24,12))
ax = sb.lineplot(x=data["ArraySize"], y=data["MergeSort"][900:1000], color='blue', label='mergeSort')
ax = sb.lineplot(x=data["ArraySize"], y=data["HybridSort"][900:1000], color='red', label='hybridSort')
ax.set(ylabel="Time (ms)")
```

```python
plt.figure(figsize=(24,12))
ax = sb.lineplot(x=data["ArraySize"], y=data["MergeSort"][:100], color='blue', label='mergeSort')
ax = sb.lineplot(x=data["ArraySize"], y=data["HybridSort"][:100], color='red', label='hybridSort')
ax.set(ylabel="Time (ms)")
```
